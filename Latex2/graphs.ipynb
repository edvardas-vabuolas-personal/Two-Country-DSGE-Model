```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, theme_bw, theme, labs, scale_x_continuous, scale_y_continuous
```

```python
x = [x for x in range(1,50)]
df = pd.DataFrame(data={'x':x})

for sigma in [1.5, 1.8, 2.1, 2.4]:
    
    y = [(x**(1-sigma))/(1-sigma) for x in x]

    df[f'$\sigma$ = {sigma}'] = y

df = pd.melt(df, id_vars=['x'])
plt = ggplot(df) + aes(x='x', y='value', colour='variable') + geom_line() + theme_bw() + theme(legend_position = "top") + labs(color="") + scale_x_continuous(name="Consumption $C$") + scale_y_continuous(name=r"Utility $\mathcal{U}$")
plt.
```
